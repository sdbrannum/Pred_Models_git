```python
#import packages
import numpy as py
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_squared_error
%matplotlib inline 
```

```python
# add headers to dataset

crime.columns = [#'crimepredict'
'state','county','community','communityname','fold','population','householdsize','racepctblack','racePctWhite','racePctAsian','racePctHisp','agePct12t21','agePct12t29','agePct16t24','agePct65up','numbUrban','pctUrban','medIncome','pctWWage','pctWFarmSelf','pctWInvInc','pctWSocSec','pctWPubAsst','pctWRetire','medFamInc','perCapInc','whitePerCap','blackPerCap','indianPerCap','AsianPerCap','OtherPerCap','HispPerCap','NumUnderPov','PctPopUnderPov','PctLess9thGrade','PctNotHSGrad'
,'PctBSorMore','PctUnemployed','PctEmploy','PctEmplManu','PctEmplProfServ','PctOccupManu','PctOccupMgmtProf','MalePctDivorce','MalePctNevMarr','FemalePctDiv','TotalPctDiv','PersPerFam','PctFam2Par','PctKids2Par','PctYoungKids2Par','PctTeen2Par','PctWorkMomYoungKids','PctWorkMom','NumIlleg','PctIlleg'
,'NumImmig','PctImmigRecent','PctImmigRec5','PctImmigRec8','PctImmigRec10','PctRecentImmig','PctRecImmig5','PctRecImmig8','PctRecImmig10','PctSpeakEnglOnly'
,'PctNotSpeakEnglWell','PctLargHouseFam','PctLargHouseOccup','PersPerOccupHous','PersPerOwnOccHous'
,'PersPerRentOccHous','PctPersOwnOccup','PctPersDenseHous','PctHousLess3BR','MedNumBR','HousVacant','PctHousOccup','PctHousOwnOcc'
,'PctVacantBoarded','PctVacMore6Mos','MedYrHousBuilt','PctHousNoPhone','PctWOFullPlumb','OwnOccLowQuart','OwnOccMedVal','OwnOccHiQuart','RentLowQ'
,'RentMedian','RentHighQ','MedRent','MedRentPctHousInc','MedOwnCostPctInc','MedOwnCostPctIncNoMtg','NumInShelters','NumStreet','PctForeignBorn'
,'PctBornSameState','PctSameHouse85','PctSameCity85','PctSameState85','LemasSwornFT','LemasSwFTPerPop','LemasSwFTFieldOps','LemasSwFTFieldPerPop'
,'LemasTotalReq','LemasTotReqPerPop','PolicReqPerOffic','PolicPerPop','RacialMatchCommPol','PctPolicWhite','PctPolicBlack','PctPolicHisp','PctPolicAsian','PctPolicMinor','OfficAssgnDrugUnits','NumKindsDrugsSeiz','PolicAveOTWorked','LandArea','PopDens','PctUsePubTrans','PolicCars','PolicOperBudg'
,'LemasPctPolicOnPatr','LemasGangUnitDeploy','LemasPctOfficDrugUn','PolicBudgPerPop'
,'ViolentCrimesPerPop']
```

```python
# check data was read in
crime.head()
```

```python
#check shape
crime.shape
```

```python
# replace question marks with NaN
crime = crime.replace('?', py.NaN)
crime.head()
```

```python
# check count of NaNs in columns

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
print(crime.isnull().sum())
```

```python
# remove state, fold, country, community, community name as i don't believe those to be valid predictors
# also drop features with an inordinate amount of NaNs
crime = crime.drop(['state', 'fold', 'county', 'community', 'communityname'], axis=1)
crime = crime.drop(['LemasSwornFT', 'LemasSwFTFieldOps', 'LemasSwFTFieldPerPop', 'LemasTotalReq', 'LemasTotReqPerPop'], axis=1)
crime = crime.drop(['PolicReqPerOffic', 'PolicPerPop', 'RacialMatchCommPol', 'PctPolicWhite', 'PctPolicBlack', 'PctPolicHisp'], axis=1)
crime = crime.drop(['PctPolicAsian', 'PctPolicMinor', 'OfficAssgnDrugUnits', 'NumKindsDrugsSeiz', 'PolicAveOTWorked', 'PolicCars'], axis=1)
crime = crime.drop(['PolicOperBudg', 'LemasPctPolicOnPatr', 'LemasGangUnitDeploy', 'PolicBudgPerPop', 'LemasSwFTPerPop'], axis=1)
crime.head()
```

```python
print(crime.isnull().sum())
```

```python
# convert otherpercap to float64, check otherspercap values since it's only missing one I don't want to drop the column
crime['OtherPerCap'] = crime['OtherPerCap'].astype(float)

print(crime['OtherPerCap'].median())
print(crime['OtherPerCap'].std())
print(crime['OtherPerCap'].mean())
print(crime['OtherPerCap'].min())
print(crime['OtherPerCap'].max())
```

```python
# fill missing OtherPerCap
crime['OtherPerCap'] = crime['OtherPerCap'].replace(py.NaN, crime['OtherPerCap'].mean())
# print(crime.isnull().sum())
```

```python
# breaking our dataset into the target and the predictors
crime.target=crime['ViolentCrimesPerPop'] 
crime.features=crime.drop(['ViolentCrimesPerPop'], axis=1)                         
print(crime.target.shape)
print(crime.features.shape)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression

# fit a linear regression model to the data
LRModel = LinearRegression(normalize=True)
LRModel.fit(crime.features, crime.target)
print(LRModel)

# make predictions
expected_LR = crime.target
predicted_LR = LRModel.predict(crime.features)

# summarize the fit of the model
print("Coef", LRModel.intercept_, LRModel.coef_)
print("MSE", mean_squared_error(expected_LR, predicted_LR))
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge

RGModel = Ridge(alpha=20)
RGModel.fit(crime.features, crime.target)
print(RGModel)
# make predictions
expected_RG = crime.target
predicted_RG= RGModel.predict(crime.features)
# summarize the fit of the model
print("Coef", RGModel.intercept_, RGModel.coef_)
print("MSE", mean_squared_error(expected_RG, predicted_RG))
```

```python
# grid search on ridge

# use a full grid over several parameters and cross validate 5 times
from sklearn.model_selection import GridSearchCV
param_grid = {"alpha": [.01,.1, .5, 1, 2]}
#param_grid={"alpha": [1,10,1]} this does a range 1 through 10 changes by a factor of 1. 
#param_grid={"alpha": [.01,1,.05]} this does a range 1 through 1 changes by a factor of .05

# run grid search
grid_search = GridSearchCV(RGModel, param_grid=param_grid,n_jobs=-1,cv=5)
grid_search.fit(crime.features, crime.target)
print("Grid Scores", grid_search.cv_results_)
print("Best", grid_search.best_params_)
```

The grid search against the ridge regression showed that an Alpha of 2 was the best fit
when given the values of [.01, .1, .5, 1, 2].  

## Ridge Regression with Cross Validation

```python
from sklearn.linear_model import RidgeCV

RGModel = RidgeCV(alphas=[.5,1,5,10,20])
RGModel.fit(crime.features, crime.target)
print(RGModel)
# make predictions
expected_RG = crime.target
predicted_RG= RGModel.predict(crime.features)
# summarize the fit of the model
print("Coef", RGModel.intercept_, RGModel.coef_)
print("MSE", mean_squared_error(expected_RG, predicted_RG))
```

## Lasso Regression with Cross Validation

```python
from sklearn.linear_model import LassoCV

LASModel = LassoCV(alphas=[.01,.1, .5, 1, 2], cv=5)
LASModel.fit(crime.features, crime.target)
print(LASModel)
# make predictions
expected_LAS = crime.target
predicted_LAS = LASModel.predict(crime.features)
# summarize the fit of the model
print("Coef", LASModel.intercept_,LASModel.coef_)
print("MSE", mean_squared_error(expected_LAS, predicted_LAS))
```

## ElasticNet Regression

```python
from sklearn.linear_model import ElasticNet

# fit a model to the data
ENModel = ElasticNet(alpha=.5)
ENModel.fit(crime.features, crime.target)
print(ENModel)
# make predictions
expected_EN =  crime.target
predicted_EN = ENModel.predict(crime.features)
# summarize the fit of the model
print("Coef", ENModel.intercept_, ENModel.coef_)
print("MSE", mean_squared_error(expected_EN, predicted_EN))
```

## Least Angle Regression

```python
from sklearn.linear_model import Lars

# fit a model to the data
LARSModel = Lars()
LARSModel.fit(crime.features, crime.target)
print(LARSModel)
# make predictions
expected_LARS =  crime.target
predicted_LARS = LARSModel.predict(crime.features)
# summarize the fit of the model
print("Coef", LARSModel.intercept_, LARSModel.coef_)
print("MSE", mean_squared_error(expected_LARS, predicted_LARS))
```

## Lasso Lars Regression

```python
from sklearn.linear_model import LassoLars

# fit a model to the data
LLModel = LassoLars(alpha=1)
LLModel.fit(crime.features, crime.target)
print(LLModel)
# make predictions
expected_LL =  crime.target
predicted_LL = LLModel.predict(crime.features)
# summarize the fit of the model
print("Coef", LLModel.intercept_, LLModel.coef_)
print("MSE", mean_squared_error(expected_LL, predicted_LL))
```

## Optimal Model

Given the above models I would personally use either Lasso or ElasticNet regression.  I would choose lasso because it had a very low MSE but also feature selected to a very narrow feature set.  ElasticNet also had a very small MSE but it appears it dropped all features so it would be difficult to determine what features were statically significant.
