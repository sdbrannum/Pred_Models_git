## Import Data

```python
auto_data = pd.read_csv('AUTO.csv', encoding = "ISO-8859-1")
auto_data.head()
```

```python
# check shape again
auto_data.shape
```

```python
auto_data.tail()
```

```python
# check for nulls
auto_data.isnull().values.any()
```

```python
# Weight is currently a string, we want it as a numeric
# remove comma, convert weight to numeric
auto_data['Weight (lbs)'] = auto_data['Weight (lbs)'].replace({',':''},regex=True)
auto_data['Weight (lbs)'] = pd.to_numeric(auto_data['Weight (lbs)'])
auto_data['Weight (lbs)'].head()
```

```python
auto_data.dtypes
```

```python
auto_data.median()
```

## Fill missing data with medians

```python
# get rows where luggage has an na just so we can verify the fill worked correctly below
nan_rows = auto_data[auto_data['Luggage (cu. ft.)'].isnull()]
nan_rows.head()
```

```python
# fill with medians
auto_data = auto_data.fillna(auto_data.median())
```

```python
# verification that the nas were filled
auto_data.isnull().values.any()
```

```python
# verification they were filled with medians
auto_data[87:89]
```

```python
auto_data.describe()
```

```python
# distribution of drive type
auto_data.groupby('Drive Type').size()
```

```python
# distribution of fuel type
auto_data.groupby('Fuel Type').size()
```

```python
# Fix Fuel Type by removing blank spaces
auto_data['Fuel Type'] = auto_data['Fuel Type'].str.strip()
auto_data.groupby('Fuel Type').size()
```

```python
# z score normalization on weight then insert next to weight
weightZ = pd.DataFrame((auto_data['Weight (lbs)'] - auto_data['Weight (lbs)'].mean())/ auto_data['Weight (lbs)'].std())
auto_data.insert(10, 'WeightZ', weightZ)
auto_data.head()
```

```python
# convert categorical to dummies into new data frame
auto_dummied = pd.get_dummies(auto_data, columns=['Drive Type', 'Fuel Type'])
auto_dummied.head()
```
